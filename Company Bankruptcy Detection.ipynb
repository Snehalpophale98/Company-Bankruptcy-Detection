# Data Modeling

```python
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTEENN
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import confusion_matrix,classification_report
import sklearn.metrics as metrics
from sklearn.model_selection import GridSearchCV


from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
X = df.drop('Bankrupt?', axis=1)
y = df['Bankrupt?']
```

```python
fs = SelectKBest(score_func=f_classif, k=int((df.shape[1]*90)/100))               #feature selection
fs
```

```python
X = fs.fit_transform(X, y)
X = pd.DataFrame(X)
y = pd.DataFrame(y)
```

```python
y.value_counts()              # Data is imbalance
```

### Balancing the data using SMOTE

```python
sm = SMOTEENN()    #oversampling
x_res , y_res = sm.fit_resample(X,y)
y_res.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_res, y_res, test_size=0.2, random_state=42)
```

### Random Forest Classifier

```python
model_RF = RandomForestClassifier(n_estimators=150,criterion='gini',random_state=100,max_depth=10,min_samples_leaf=11)
```

```python
model_RF.fit(x_train , y_train)
```

```python
y_predictRF = model_RF.predict(x_test)
```

```python
model_RF.score(x_test,y_test) 
```

```python
print(classification_report(y_test,y_predictRF))
```

```python
confusion_matrixRF = metrics.confusion_matrix(y_test,y_predictRF)
print(confusion_matrixRF)
```

```python
sns.heatmap(confusion_matrixRF, annot=True, fmt = 'd', cbar = False)
```
